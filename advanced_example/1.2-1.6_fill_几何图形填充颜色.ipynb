## 1.1.2 不规则多边形的颜色填充
* matplotlib.pyplot.fill()
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill.html

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,500)
y = np.sin(x)

plt.fill(x,y,color="cornflowerblue",alpha=0.4)
# alpha=0.4 透明度

plt.plot(x,y,color="cornflowerblue",alpha=0.8)
plt.plot([x[0],x[-1]],[y[0],y[-1]],color="cornflowerblue",alpha=0.8)

plt.xlim(0,2*np.pi)
plt.ylim(-1.1,1.1)

plt.show()
```

## 1.2 交叉曲线的颜色填充
* matplotlib.pyplot.fill_between(x,y1,y2)
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html?highlight=fill_between#matplotlib.pyplot.fill_between

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,500)
y1 = np.sin(2*np.pi*x)
y2 = 1.1*np.sin(3*np.pi*x) #最大值 1.1

fig,ax = plt.subplots(3,1,sharex="all")

# x轴全部填充，y轴中 "between y2 and 0" 之间填充
ax[0].fill_between(x,0,y2,alpha=0.5)
ax[0].set_ylim(-1.2,1.2)

# "between y2 and 1.1"
ax[1].fill_between(x,y2,1.1,alpha=0.5)
ax[1].set_ylim(-1.2,1.2)

# "between y1 and y2"
ax[2].fill_between(x,y1,y2,alpha=0.5)
ax[2].set_xlim(0,2)
ax[2].set_ylim(-1.2,1.2)

plt.show()
```

## 1.3.1 水平方向的交叉曲线的颜色填充方法
* matplotlib.pyplot.fill_between(x,y1,y2,where=y2>=y1,interpolate=True,
                facecolor="cornflowerblue",alpha=0.7)
* where: 条件控制
* interpolate: True, 交叉点附近公共交叉区域使用指定颜色填充;#好像并没有用
* facecolor: 背景填充颜色
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html?highlight=fill_between#matplotlib.pyplot.fill_between

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,500)
y1 = np.sin(2*np.pi*x)
y2 = 1.1*np.sin(3*np.pi*x)

fig = plt.figure()

ax = fig.add_subplot(111)

# plot y1 and plot y2
ax.plot(x,y1,color="k",lw=1,ls="-")
ax.plot(x,y2,color="k",lw=1,ls="-")

# "where y1 <= y2"
ax.fill_between(x,y1,y2,where=y2>=y1,interpolate=True,
                facecolor="cornflowerblue",alpha=0.7)


# where y1>= y2
ax.fill_between(x,y1,y2,where=y2<=y1,interpolate=True,
                facecolor="darkred",alpha=0.7)

ax.set_xlim(0,2)
ax.set_ylim(-1.2,1.2)

ax.grid(ls=":",lw=1,color="gray",alpha=0.5)

plt.show()
```

## 1.3.2 垂直方向的交叉曲线的颜色填充方法
* fill_betweenx(y,x1,x2,where=x2>=x1,facecolor="cornflowerblue",alpha=0.7)

```python
import matplotlib.pyplot as plt
import numpy as np

# 方向根据，x,y轴的填入有关
y = np.linspace(0,2,500)
x1 = np.sin(2*np.pi*y)
x2 = 1.1*np.sin(3*np.pi*y)

fig = plt.figure()

ax = fig.add_subplot(111)

# plot x1 and plot x2
ax.plot(x1,y,color="k",lw=1,ls="-")
ax.plot(x2,y,color="k",lw=1,ls="-")

# 方向根据，x,y轴的填入有关

# "where x1 <= x2"
ax.fill_betweenx(y,x1,x2,where=x2>=x1,facecolor="cornflowerblue",alpha=0.7)

# where x1>= x2
ax.fill_betweenx(y,x1,x2,where=x2<=x1,facecolor="darkred",alpha=0.7)

ax.set_xlim(-1.2,1.2)
ax.set_ylim(0,2)

ax.grid(ls=":",lw=1,color="gray",alpha=0.5)

plt.show()
```

## 1.4 交叉间断型曲线的颜色填充
# numpy
* np.ma.masked_greater(y2,1.0) 
* ma.masked_greater数组中，大于1.0的数值做掩饰处理，图中就不会填充颜色了

```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots(1,2)

# subplot(121) data
x = np.linspace(0,2,500)
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(3*np.pi*x)

y2 = np.ma.masked_greater(y2,1.0) #数组中，大于1.0的数值做掩饰处理，图中就不会填充颜色了

# plot y1 and plot y2
ax[0].plot(x,y1,color="k",lw=1,ls="-")
ax[0].plot(x,y2,color="k",lw=1,ls="-")

# "where y1 <= y2"
ax[0].fill_between(x,y1,y2,where=y2>=y1,facecolor="cornflowerblue",alpha=0.7)

# where y1>= y2
ax[0].fill_between(x,y1,y2,where=y2<=y1,facecolor="darkred",alpha=0.7)

ax[0].set_xlim(0,2)
ax[0].set_ylim(-1.2,1.2)

ax[0].grid(ls=":",lw=1,color="gray",alpha=0.5)

# subplot(122) data
y = np.linspace(0,2,500)
x1 = np.sin(2*np.pi*y)
x2 = 1.2*np.sin(3*np.pi*y)

x2 = np.ma.masked_greater(x2,1.0)

# plot x1 and plot x2
ax[1].plot(x1,y,color="k",lw=1,ls="-")
ax[1].plot(x2,y,color="k",lw=1,ls="-")

# "where x1 <= x2"
ax[1].fill_betweenx(y,x1,x2,where=x2>=x1,facecolor="cornflowerblue",alpha=0.7)

# where x1>= x2
ax[1].fill_betweenx(y,x1,x2,where=x2<=x1,facecolor="darkred",alpha=0.7)

# x,y轴长度限制
ax[1].set_xlim(-1.2,1.2)
ax[1].set_ylim(0,2)

# 网格线设置
ax[1].grid(ls=":",lw=1,color="gray",alpha=0.5)

plt.show()
```
