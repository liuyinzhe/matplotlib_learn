```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.5,2*np.pi,20)
#正态分布中随机选取20个数值 (0,1]
y = np.random.randn(20) 

markerline,stemlines,baseline = plt.stem(x,y,linefmt="-.",markerfmt="D",basefmt="-")

#plt.setp() 添加设置
plt.setp(markerline,color="chartreuse",marker="D")
plt.setp(stemlines,linestyle="-.")
#baseline.set_linewidth(2)
plt.setp(baseline,linewidth=2)

# 保存图片
plt.savefig(fname='stem_棉棒图.png', dpi=300)

plt.show()
```
