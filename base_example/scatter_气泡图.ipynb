```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

a = np.random.randn(100) # 正态分布(0,1] 中取随机100个数字
b = np.random.randn(100) # 正态分布(0,1] 中取随机100个数字

# colormap:RdYlBu
plt.scatter(a,b,s=np.power(10*a+20*b,2),
            c=np.random.rand(100),
            cmap=mpl.cm.RdYlBu,
            marker="o") #marker O 圆形

# 保存图片
plt.savefig(fname='scatter_气泡图.png', dpi=300)

plt.show()
```
