```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()

num = 50

# new sample
'''
np.random.rand(4,3) #4行3列
生成一个shape为[4,3]的array，array中每个元素都是一个[0,1)之间的随机数
'''
sample = 10*np.random.rand(num,2)
var1 = sample[:,0]
var2 = sample[:,1]

# threshold value
td = 12

# discriminant function
df = 2*var1+var2
'''
np.ma.masked_where(condition,var)
根据条件condition，对var[dateframe] 中对应数值进行判断，符合条件的替换为--；
var 中不符合条件的保持原来的数字
'''
cates11 = np.ma.masked_where(df>=td,var1)
cates12 = np.ma.masked_where(df>=td,var2)

cates21 = np.ma.masked_where(df<=td,var1)
cates22 = np.ma.masked_where(df<=td,var2)

#绘图
'''
s size 大小
s : scalar or array_like, shape (n, ), optional
The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.

marker 形状
c  color 颜色
'''
ax.scatter(var1,var2,s=cates11*50,marker="s",c=cates11)
ax.scatter(var1,var2,s=cates21*50,marker="o",c=cates21)

# 绘制蓝色直线
ax.plot(var1,-2*var1+12,lw=1,color="b",alpha=0.65)

ax.axis([-1,11,-1,11])


# 保存图片
plt.savefig(fname='scatter_分组气泡图.png', dpi=300)

plt.show()
```
